Install and Read in Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import itertools
import statsmodels.api as sm
from matplotlib.dates import DateFormatter

plt.style.use('fivethirtyeight')

# load data
df = pd.read_csv("GlobalWeatherRepository.csv")
df.info()
```

Data Cleaning & Preprocessing


```python
# Select numerical columns
numerical_cols = [
    "temperature_celsius", "precip_mm", "wind_kph", "pressure_mb", "humidity", "uv_index"
]

# Plot boxplots to visualize outliers
plt.figure(figsize=(15, 8))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
df = df.drop_duplicates()
```

```python
# Filtering data for a specific country
country_name = 'Ukraine'
df_country = df[df['country'] == country_name].copy()

# Selecting necessary columns and preparing them
df_country = df_country[['last_updated', 'temperature_celsius']].copy()
df_country["last_updated"] = pd.to_datetime(df_country["last_updated"])
df_country['temperature_celsius'] = df_country['temperature_celsius'].astype(int)

# Removing anomalies using the IQR method
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = df[column].clip(lower_bound, upper_bound)
    return df

df_country = remove_outliers(df_country, 'temperature_celsius')
df_country = df_country.drop_duplicates(subset=['last_updated'])

# Setting the index and aggregating data by day
df_country = df_country.set_index('last_updated')
daily_data = df_country.resample('D').mean()

# Filling missing values using linear interpolation
daily_data['temperature_celsius'] = daily_data['temperature_celsius'].interpolate(method='linear')
daily_data = daily_data.sort_index()

# Plotting the graph
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(daily_data.index, daily_data['temperature_celsius'], marker='o', linestyle='-')
ax.set_xlim(daily_data.index.min(), daily_data.index.max())
ax.set_xlabel('Date')
ax.set_ylabel('Temperature (C)')
ax.set_title(f'Daily Temperature Chart for {country_name}')

# Formatting the time axis
date_form = DateFormatter("%Y-%m-%d")
ax.xaxis.set_major_formatter(date_form)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Train Model

```python
p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))

seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]

print('Examples of parameter combinations for Seasonal ARIMA...')
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[1]))
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[2]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[3]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[4]))
```

```python
warnings.filterwarnings("ignore")

for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            model = sm.tsa.statespace.SARIMAX(daily_data,
                                            order=param,
                                            seasonal_order=param_seasonal,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)
            results = model.fit()
            print(f'ARIMA{param}x{param_seasonal}12 - AIC:{results.aic}')
        except:
            continue
```

```python
best_model = sm.tsa.statespace.SARIMAX(daily_data,
                                order=(1,1,1),
                                seasonal_order=(0,1,1,12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)
results = best_model.fit()
print(results.summary().tables[1])
```

```python
results.plot_diagnostics(figsize=(15, 12))
plt.show()
```

Forecast

```python
pred = results.get_prediction(start=pd.to_datetime('2025-01-21'), dynamic=False)
pred_ci = pred.conf_int()
pred_ci
```

```python
y_forecasted = pred.predicted_mean
y_truth = daily_data['2025-01-21':]
y_truth['Pred_Temp'] = y_forecasted
y_truth
```

```python
ax = daily_data['2025-01-21':].plot(label='Observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7)

ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)
ax.set_xlabel('Date')
ax.set_ylabel('Temperature')
plt.legend()

plt.show()
```

```python
mse = ((y_truth['temperature_celsius'] - y_truth.Pred_Temp)**2).mean()
print('Mean Squared Error = ', round(mse,2))
```

```python
rmse = np.sqrt(mse)
print('Root Mean Squared Error = ', rmse)
```

```python
pred_uc = results.get_forecast(steps=10)
pred_ci = pred_uc.conf_int()
```
